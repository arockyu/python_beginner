$$a_{n}=\frac{8}{\pi}\frac{1}{n^{2}}\sin(\frac{n\pi}{2})$$
$$f(x,N)=\sum_{n=1}^{N}a_n\sin{nx}$$

```python
from matplotlib import pyplot as plt
import numpy as np

def func(xx,nn):
    ret=0.
    for ii in range(1,1+nn):
        an=8./np.pi/(ii**2)*np.sin(ii*np.pi/2.)
        ret+=an*np.sin(ii*xx)
    return ret 
      
rows=3
columns=6

t=np.linspace(0,2,100)
f=2
omega=2.*np.pi*f

fig, axs = plt.subplots(rows, columns,figsize=[10.8,4.8]) 

for i in range(rows):
    for j in range(columns):
        n=(i*columns+j)+1
        axs[i][j].plot(t,func(omega*t,n),label="N="+str(n))
        axs[i][j].legend()
```
