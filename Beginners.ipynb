$$a_{n}=\frac{1}{2n-1}$$
$$f(x,N)=\sum_{n=1}^{N}a_n\sin{(2n-1)x}$$

```python
from matplotlib import pyplot as plt
import numpy as np

def func(xx,nn):
    ret=0.
    for ii in range(1,1+nn):
        an=1.0/(2*ii-1)
        ret+=an*np.sin((2*ii-1)*xx)
    return ret 
      
rows=3
columns=6

t=np.linspace(-1,1,100)
f=1
omega=2.*np.pi*f

fig, axs = plt.subplots(rows, columns,figsize=[10.8,4.8]) 

for i in range(rows):
    for j in range(columns):
        n=(i*columns+j)+1
        axs[i][j].plot(t,func(omega*t,n),label="2N-1="+str(2*n-1))
        axs[i][j].legend()
```
